```python
import numpy as np

# 변수 설정
t = 10  # 시간 단계 수
sample = 10000  # 샘플 수

# 상태 전이 확률 행렬 P
P = np.array([[0.7, 0.2, 0.1],
              [0.3, 0.65, 0.05],
              [0.15, 0.35, 0.5]])

# 각 시간 단계에서 상태의 빈도를 저장할 리스트 초기화
Xts_counts = [[] for _ in range(t)]

# 샘플 수만큼 반복하여 시뮬레이션
for i in range(sample):
    X0 = 1  # 초기 상태 설정
    Xt = X0  # 현재 상태를 초기 상태로 설정
    for j in range(t):
        # 다음 상태를 현재 상태의 전이 확률에 따라 결정
        next = np.random.choice([1, 2, 3], p=P[Xt-1])
        Xt = next  # 현재 상태를 다음 상태로 업데이트
        Xts_counts[j].append(next)  # 해당 시간 단계에 다음 상태를 추가

# 각 시간 단계별로 상태 빈도 출력
for j in range(t):
    print(Xts_counts[j].count(1), Xts_counts[j].count(2), Xts_counts[j].count(3))
```
