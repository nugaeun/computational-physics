```python
import numpy as np
```

```python
# 1

# 변수 설정
n = 1000000  # 시행 수
N = 50  # 각 시행의 단계 수
p = 0.4  # 1이 나올 확률
step_list = [-1, 1]  # 한 단계에서의 이동 가능 값 (-1 또는 1)

# n*N 크기의 배열 생성 후 각 요소에 step_list에서 랜덤하게 선택된 값 할당
# 각 요소는 -1 또는 1이며, -1이 나올 확률은 1-p, 1이 나올 확률은 p
Xn = np.random.choice(step_list, size=(n, N), p=[1-p, p])

# 각 시행 결과의 합 계산 (axis=1 : 각 행의 요소들을 합하여 열 벡터로 반환)
Xn_sum = np.sum(Xn, axis=1)

# 합이 15 이상인 경우의 수 계산
over15 = np.sum(Xn_sum > 15)

# 15 이상의 값들이 나올 확률을 계산
prob = over15 / n
print(prob)  # 계산된 확률 출력
```

```python
# 2

# 변수 설정
n = 1000000  # 시행 수
N = 50  # 각 시행의 단계 수

p = 0.4  # 첫 번째 이동이 1일 확률
q = 0.65  # 두 번째 이동이 1일 확률
step_list = [-1, 1]  # 한 단계에서의 이동 가능 값 (-1 또는 1)

# 오른쪽 방향의 확률과 왼쪽 방향의 확률 계산
right = p / q
left = (1 - p) / (1 - q)

# n*N 크기의 배열 생성 후 각 요소에 step_list에서 랜덤하게 선택된 값 할당
# 각 요소는 -1 또는 1이며, -1이 나올 확률은 1-q, 1이 나올 확률은 q
Xn = np.random.choice(step_list, size=(n, N), p=[1 - q, q])

# 각 시행의 결과를 합하여 15 이상인 경우의 인덱스 찾기
over15_index = np.sum(Xn, axis=1) > 15

# 15 이상인 시행들만 선택
selected_Xn = Xn[over15_index]

# 선택된 시행들의 결과가 1이면 right 값을, -1이면 left 값을 가지는 배열을 생성하여 모두 곱함
# 즉, 각 선택된 시행들의 전체 확률을 계산
probs = np.prod(np.where(selected_Xn == 1, right, left), axis=1)

# 전체 확률 계산하여 출력
fr = np.sum(probs)
print(fr / n)
```
