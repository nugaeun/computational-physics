```python
# MC1의 single realization
import numpy as np

t = 100
P = np.array([[0.7, 0.2, 0.1],
              [0.3, 0.65, 0.05],
              [0.15, 0.35, 0.5]])

initial = 1
current = initial
Xt = []
for _ in range(t):
  # 초기값이 1이므로 현재 상태가 1일 때, 다음 상태를 선택
  # 선택할 수 있는 후보는 1, 2, 3
  # 각 후보의 확률은 전이 확률 행렬에서 현재 상태에 해당하는 행의 값이다.
  next = np.random.choice([1, 2, 3], p=P[current-1])
  current = next
  Xt.append(current)

print(f'각 상태별 발생 횟수:\n상태 1:{Xt.count(1)}\n상태 2:{Xt.count(2)}\n상태 3:{Xt.count(3)}')
```
