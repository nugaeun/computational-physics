```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 예제 5.3.1

# 변수 설정
N = 10000  # 샘플 수
a = 0  # 하한값
b = 10  # 상한값

# 함수 p(x): 균등분포 함수
def p(x):
    return np.where((a<x)&(x<b), 1/(b-a), 0)

# 함수 q(x): 지수분포 함수
def q(x):
    return np.exp(-x)/(1-np.exp(-10))

# 함수 g(x): 목표 함수
def g(x):
    return np.exp(-0.5*x**2)

# 함수 w(x): 중요도 샘플링 가중치 함수
def w(x):
    return p(x)/q(x)

# 균등분포를 사용한 샘플링
xi = np.random.uniform(a, b, N)
Ep = (1/N)*(np.sum(g(xi)))  # Ep 계산
print(Ep)

# 지수분포를 사용한 샘플링
u = np.random.uniform(0, 1, N)
xi = -np.log(1-(1-np.exp(-10))*u)  # 지수분포 샘플 생성
Eq = (1/N)*(np.sum(g(xi)*w(xi)))  # Eq 계산
print(Eq)
```

```python
# 예제 5.3.2

# 변수 설정
N = 10000  # 샘플 수
z = 3  # 기준값

# 함수 g(x): x가 z보다 큰 경우 1, 그렇지 않으면 0
def g(x):
    return np.where((x>z), 1, 0)

# 함수 p(x): 표준 정규분포 함수
def p(x):
    return 1/(2*np.pi)**(1/2)*np.exp(-x**2/2)

# 함수 q(x): 평균이 z인 정규분포 함수
def q(x):
    return 1/(2*np.pi)**(1/2)*np.exp(-(x-z)**2/2)

# 함수 w(x): 중요도 샘플링 가중치 함수
def w(x):
    return p(x)/q(x)

# 표준 정규분포를 사용한 샘플링
xi = np.random.normal(0, 1, N)
Ep = (1/N)*(np.sum(g(xi)))  # Ep 계산
print(Ep)

# 평균이 z인 정규분포를 사용한 샘플링
xi = np.random.normal(z, 1, N)
Eq = (1 / N) * (np.sum(g(xi) * w(xi)))  # Eq 계산
print(Eq)
```
