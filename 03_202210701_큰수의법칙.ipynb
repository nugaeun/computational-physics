```python
import numpy as np
import matplotlib.pyplot as plt

def LLN(N):
    # N을 100으로 나눈 값을 정수로 변환하여 n 설정
    n = int(N / 100)

    # 0과 1 사이의 균등 분포를 따르는 N개의 난수 생성
    x = np.random.uniform(0, 1, N)

    # 생성된 난수를 (n, N/n) 형태의 배열로 재구성
    y = np.reshape(x, (n, int(N/n)))

    # 각 행의 평균값을 구함
    ys = y.mean(axis=1)

    # 각 행의 평균값의 합을 구함
    s = ys.sum()

    # 각 행의 평균값의 평균을 구함
    m = s / int(N/n)

    # 편차(표준편차)를 계산함
    std_dev = np.std(ys)

    # 그래프 설정
    plt.figure(figsize=(4, 2))

    # 표준편차의 히스토그램을 그림
    plt.hist(ys, bins=100)

    # 평균값을 나타내는 빨간색 점선
    plt.axvline(x=ys.mean(), color='r', linestyle='--', linewidth=1, label='Mean')

    # 평균값 + 표준편차를 나타내는 초록색 점선
    plt.axvline(x=ys.mean() + std_dev, color='g', linestyle='--', linewidth=1)

    # 평균값 - 표준편차를 나타내는 초록색 점선
    plt.axvline(x=ys.mean() - std_dev, color='g', linestyle='--', linewidth=1)

    # 그래프 표시
    plt.show()
```

```python
LLN(100000)
LLN(1000000)
LLN(10000000)
LLN(100000000)
```
