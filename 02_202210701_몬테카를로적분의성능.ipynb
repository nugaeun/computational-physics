```python
import random as rd
import numpy as np
import math

# 균등 분포에 대한 함수
def g_uni(x):
    return 3*x**2

def f_uni(x):
    return 1

# 베타 분포에 대한 함수 (매개변수 a와 b)
def g_beta(x):
    return 3*x**2

def f_beta(x, a, b):
    return (math.gamma(a+b)/(math.gamma(a)*math.gamma(b)))*x**(a-1)*(1-x)**(b-1)

# f(y) = 2y 분포에 대한 함수
def g_fy(x):
    return 3*x**2

def f_fy(y):
    return 2*y

# 샘플 수와 실험 횟수
N = 10000
n = 10
a, b = 2, 2

# 균등 분포에 대한 기대값과 분산 계산
E_uni_list = []

for _ in range(n):
    E_uni = 0
    for _ in range(N):
        x = rd.uniform(0, 1)
        g_value = g_uni(x)/f_uni(x)
        E_uni += g_value
    E_uni /= N
    E_uni_list.append(E_uni)

E_uni_mean = sum(E_uni_list)/n
E_uni_sigma = sum((E - E_uni_mean)**2 for E in E_uni_list)
var_uni = 1/(n-1) * E_uni_sigma
print("1. uniform:")
print("평균:", E_uni_mean)
print("표본분산:", var_uni)

# 베타 분포에 대한 기대값과 분산 계산
E_beta_list = []

for _ in range(n):
    E_beta = 0
    for _ in range(N):
        x = np.random.beta(a, b)
        g_value = g_beta(x)/f_beta(x, a, b)
        E_beta += g_value
    E_beta /= N
    E_beta_list.append(E_beta)

E_beta_mean = sum(E_beta_list)/n
E_beta_sigma = sum((E - E_beta_mean)**2 for E in E_beta_list)
var_beta = 1/(n-1) * E_beta_sigma
print("\n2. beta(a,b):")
print("평균:", E_beta_mean)
print("표본분산:", var_beta)

# f(y) = 2y 분포에 대한 기대값과 분산 계산
E_fy_list = []

for _ in range(n):
    E_fy = 0
    for _ in range(N):
        x = rd.uniform(0, 1)
        x = x**(1/2)
        g_value = g_fy(x)/f_fy(x)
        E_fy += g_value
    E_fy /= N
    E_fy_list.append(E_fy)

E_fy_mean = sum(E_fy_list)/n
E_fy_sigma = sum((E - E_fy_mean)**2 for E in E_fy_list)
var_fy = 1/(n-1) * E_fy_sigma
print("\n3. f(y) = 2y:")
print("평균:", E_fy_mean)
print("표본분산:", var_fy)

# 분산의 대소 관계 비교
print("\n대소관계 비교:")
if var_uni > var_beta and var_uni > var_fy:
    print("uniform 분산이 가장 큽니다.")
elif var_beta > var_uni and var_beta > var_fy:
    print("beta 분산이 가장 큽니다.")
else:
    print("f(y) = 2y 분산이 가장 큽니다.")

if var_uni < var_beta and var_uni < var_fy:
    print("uniform 분산이 가장 작습니다.")
elif var_beta < var_uni and var_beta < var_fy:
    print("beta 분산이 가장 작습니다.")
else:
    print("f(y) = 2y 분산이 가장 작습니다.")
```
